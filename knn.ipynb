```python
import pandas as pd
```

```python
dataset = pd.read_excel('dataset.xlsx')
```

```python
dataset.head()
```

```python
# 2:5 => use avg_price
# 2:4 => not use avg_price
x = dataset.iloc[:, 2:5].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# le = LabelEncoder()

# y_train = le.fit_transform(y_train)
# y_test = le.fit_transform(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
```

```python
f1s = []
for i in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    f1 = f1_score(y_test, pred_i, pos_label='Laku')
    f1s.append({i: f1})
f1s
```

```python
best_f1 = f1s[0]
for f1 in f1s:
    if list(best_f1.values())[0] < list(f1.values())[0]:
        best_f1 = f1
best_f1
```

```python
best_k = list(best_f1.keys())[0]
best_k
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = knn.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import classification_report, recall_score, precision_score
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
precision = precision_score(y_test, y_pred, pos_label='Laku')
recall = recall_score(y_test, y_pred, pos_label='Laku')
f1 = f1_score(y_test, y_pred, pos_label='Laku')
print(precision)
print(recall)
print(f1)
```
